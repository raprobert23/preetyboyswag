# Prepare the data from the filefolders

A la hora de preparar los datos, hemos decidido colocar varios de los trozos de video en distintas carpetas, cada una recibe el nombre de la película desde la que se obtienen dichos fragmentos distorsionados. Cada una de estas carpetas se encuentra en una, que a su vez, abarca a todas las que se contiene una de las resoluciones a estudaiar.

En nuestro caso hemos decidido estudiar dos resoluciones diferentes, ajustadas a una relación de aspecto 16:9 (1920 X 1080, 1280 x 720). Estas se corresponden a Full HD y HD básico. 
Dentro de cada una de las resoluciones se encuentran los subfolders con las películas modificadas para cada una de las mismas,  que además, en las que hemos decidido evaluar la calidad de los chunks de video (8MB, 5MB, 3MB, 1MB, 0.5MB).

Hemos planteado por un lado el tener por un lado los archivos de video distorsionados (los cuales se han comparado con los archivos de referencia obtenidos a partir de la película original) y por otro lado los archovos .CSV donde se encuentra la notas obtenidas a partir de comparar los chunks de video con la herramienta de VMAF.

Como ya hemos explicado con anterioridad VMAF es capaz de comparar cada uno de los frames de un video de referencia (a) y un video distorsionado (b). De esta manera obtener una puntuación que pretende ajustarse a como la visión humana detectaría las diferencias de calidad entre ambos chunks de video. La puntuación se otorga del 0 al 100, pero nosotros hemos decidido interpolarlo entre 0 y 5 estrellas.

Donde el 0 es un stream totalmente distorsionado (pantalla en negro), y el 5 sería el mismo video de referencia. Si la referencia es peor, respecto al distorsionado, VMAF detecta que la salida del el bitrate de la distorsión es mejor que el de referencia, otorgando una puntuación ajustada a la misma.

A partir de un script en Python, hemos conseguido realizar cada una de estas llamadas al ejecutable de VMAF, alimentando los parámetros del mismo, con los videos de referencia y los videos distorsionados generados con el FFmpeg



```python
import os
import numpy as np
import re
```

```python
#Cargamos cada uno de los directorios donde se encuentran los chunks de video (20) multiplicado por cada una de las 5
#bitrates seleccionados para ser objeto de análisis (8,5,3,1,0.5 MB). Y luego todo estos para cada uan de las resoluciones
#1920 y 1280.

def sorted_alphanumeric(data):
    convert = lambda text: int(text) if text.isdigit() else text.lower()
    alphanum_key = lambda key: [ convert(c) for c in re.split('([0-9]+)', key) ] 
    return sorted(data, key=alphanum_key)

#Por algún motivo python de forma nativa considera que al seleccionar y ordenar una lista del estilo: (1 ,23,40,5...)
#en vez de realizar el orden numerico lógico ascendente (1,5,23,40), lo hace tomando como referencia el primer valor que lee
#dejandolo tal que asi: (1,23,40,5). Asi que hemos de realizar un script que lo ordene tomando com referencia el parámetro
#total numérico (del 0 al 20 en nuestro caso)

set1 = sorted_alphanumeric(os.listdir(r'D:\ffmpeg\directorios\distorted-mp4\1920X1080\avatar'))
set2 = sorted_alphanumeric(os.listdir(r'D:\ffmpeg\directorios\distorted-mp4\1280X720\avatar'))
set3 = sorted_alphanumeric(os.listdir(r'D:\ffmpeg\directorios\distorted-mp4\1920X1080\avengers'))
set4 = sorted_alphanumeric(os.listdir(r'D:\ffmpeg\directorios\distorted-mp4\1280X720\avengers'))
set5 = sorted_alphanumeric(os.listdir(r'D:\ffmpeg\directorios\distorted-mp4\1920X1080\deadpool'))
set6 = sorted_alphanumeric(os.listdir(r'D:\ffmpeg\directorios\distorted-mp4\1280X720\deadpool'))
set7 = sorted_alphanumeric(os.listdir(r'D:\ffmpeg\directorios\distorted-mp4\1920X1080\vaiana'))
set8 = sorted_alphanumeric(os.listdir(r'D:\ffmpeg\directorios\distorted-mp4\1280X720\vaiana'))

print(set2)
```
