```python
def scaleDownVideos(videos):
    for video in videos:
        folder = video.split()[0]
        #print(folder)
        path = 'D:/ffmpeg/directorios/experiments/scaled/'+folder
        #print(path)
        os.system('ffmpeg -i '+video+' -qp 1 -vf scale=320:240 -strict -2 '+path+'')

def extractImages(videos):
    os.chdir(r'D:\ffmpeg\directorios\experiments\scaled')
    videos = glob.glob(r'*.mp4')
    videos.sort(key = len)
    for video in videos:
        folder = video.split()[0]
        #print(folder)
        path = 'D:/ffmpeg/directorios/experiments/images/'+folder
        print(path)
        os.system('ffmpeg -i '+video+' -vf fps=10 '+path+'_%d.png')
```
